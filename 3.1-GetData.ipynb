# We use names-dataset python package to get country-wise popular names. The package uses the massive Facebook dump of 533M users.

```python
from names_dataset import NameDataset
import pandas as pd
```

```python
DATA_DIR = "data/"
```

```python
nd = NameDataset()
```

```python
countries = nd.get_country_codes(alpha_2=False)
countries
```

Let us focus on the following countries only

```python
country_names = {
    "India": "india",
    "United States": "usa",
    "United Kingdom": "uk",
    "Nigeria": "nigeria",
    "United Arab Emirates": "uae",
    "Morocco": "morocco",
    "Netherlands": "netherlands",
    "Peru": "peru",
    "Germany": "germany",
    "France": "france",
    "Italy": "italy",
    "Bangladesh": "bangladesh",
    "Singapore": "singapore",
    "Hong Kong": "hongkong",
    "South Africa": "southafrica",
    "Spain": "spain",
    "Mexico": "mexico",
    "Malaysia": "malaysia",
    "Canada": "canada",
    "Indonesia": "indonesia",
}
```

Fetching lastnames and gender-wise firstnames for each country

```python
data = {}
for country in countries:
    if country.name not in country_names:
        continue
    male_first_names = nd.get_top_names(n=100000, use_first_names=True, gender="M",
                            country_alpha2=country.alpha_2)[country.alpha_2]["M"]
    female_first_names = nd.get_top_names(n=100000, use_first_names=True, gender="F",
                            country_alpha2=country.alpha_2)[country.alpha_2]["F"]
    last_names = nd.get_top_names(n=100000, use_first_names=False,
                                  country_alpha2=country.alpha_2)[country.alpha_2]
    data[country_names[country.name]] = [male_first_names, female_first_names, last_names]
```

We only want to use the names which contain only the english characters: a-z

```python
dataset = []
for country in data:
    for name in data[country][0]:
        if name.isascii() and name.isalpha() and len(name) > 0:
            dataset.append([country, "firstname", "male", name.lower()])
    for name in data[country][1]:
        if name.isascii() and name.isalpha() and len(name) > 0:
            dataset.append([country, "firstname", "female", name.lower()])
    for name in data[country][0]:
        if name.isascii() and name.isalpha() and len(name) > 0:
            dataset.append([country, "lastname", "neutral", name.lower()])
```

```python
df = pd.DataFrame(dataset, columns=["country", "nametype", "gender", "name"])
df.to_csv(DATA_DIR + "dataset.csv", index=False)
df
```
